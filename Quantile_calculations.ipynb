## Water bodies and rivers quantiles
For presentation purposes I will run this code with relatively high grid step. (The values I got in run_main are more precise, but some calculations took 50 minits.)

```python
import geopandas as gpd
from shapely.geometry import Point
import numpy as np

# Load the water bodies data in Python
water_bodies = gpd.read_file("water_bodies.geojson")

# Load the rivers data in Python
rivers = gpd.read_file("rivers.geojson")
```

```python
# Load the border data in Python
border = gpd.read_file("border.geojson")
border = border.drop('NAZ_STAT', axis=1)
print(border.head())
```

```python
# Make grid points within the Czech Republic boundary

# Assuming 'czech_boundary' is a GeoDataFrame representing the boundary of Czech Republic
minx, miny, maxx, maxy = border.total_bounds

# Create a regular grid of points
grid_step = 0.1  # Adjust the grid step as needed
x, y = np.meshgrid(np.arange(minx, maxx, grid_step), np.arange(miny, maxy, grid_step))
points = gpd.GeoDataFrame(geometry=[Point(x, y) for x, y in zip(x.flatten(), y.flatten())])

# Make sure the grid points are within the Czech Republic boundary
points = gpd.sjoin(points, border, how="inner")
```

```python
# Calculate the proximities to water bodies
points['proximity_to_water_bodies'] = points['geometry'].apply(
    lambda point: water_bodies['geometry'].distance(point).min()
)

# Calculate the proximities to rivers
points['proximity_to_rivers'] = points['geometry'].apply(
    lambda point: rivers['geometry'].distance(point).min()
)
```

```python
# Get proximity quantiles

# Assuming you have a 'proximity_to_rivers' column in your 'points' GeoDataFrame
proximity_values_water_bodies = points['proximity_to_water_bodies'].to_numpy()

# Flatten and remove NaN values
proximity_values_water_bodies = proximity_values_water_bodies[~np.isnan(proximity_values_water_bodies)]

# Calculate quantiles
proximity_quantiles_water_bodies = np.percentile(proximity_values_water_bodies, [10, 20, 30, 40, 50, 60, 70, 80, 90])

# Assuming you have a 'proximity_to_rivers' column in your 'points' GeoDataFrame
proximity_values_rivers = points['proximity_to_rivers'].to_numpy()

# Flatten and remove NaN values
proximity_values_rivers = proximity_values_rivers[~np.isnan(proximity_values_rivers)]

# Calculate quantiles
proximity_quantiles_rivers = np.percentile(proximity_values_rivers, [10, 20, 30, 40, 50, 60, 70, 80, 90])

# Print results
print(proximity_quantiles_water_bodies)
print(proximity_quantiles_rivers)
```

## Forests quantiles
For presentation purposes I will run this code with relatively high grid step. (Same as before.)

```python
# Load the forests data in Python
forests = gpd.read_file("forests.geojson")
```

```python
def forested_area_percentage(coord, forests, buffer_radius):
    """
    Calculate the percentage of forested area in a circular buffer around the coordinates.

    Parameters:
        coord (tuple): Pair of coordinates as (longitude, latitude).
        forests (GeoDataFrame): GeoDataFrame representing forest polygons.
        buffer_radius (float): Radius of the circular buffer in degrees.

    Returns:
        percentage (float): Percentage of forested area in the buffer.
    """
    # Create a Shapely Point geometry from the coordinates
    point = Point(coord)

    # Create a circular buffer around the point
    buffer_zone = point.buffer(buffer_radius)

    # Calculate the intersection of the buffer zone with the forest polygons
    intersection = forests.intersection(buffer_zone)

    # Calculate the total area of the buffer zone
    total_buffer_area = buffer_zone.area

    # Calculate the total area of the forested region within the buffer
    forested_area = intersection.area.sum()

    # Calculate the percentage of forested area in the buffer
    percentage = (forested_area / total_buffer_area) * 100

    return percentage
```

```python
# Assuming 'buffer_radius' is the desired radius in degrees
buffer_radius = 0.01  # Adjust as needed

# Assuming you have a 'geometry' column in your 'points' GeoDataFrame

# Calculate the percentage of forested area around each point
points['forest_percentage'] = points['geometry'].apply(
    lambda point: forested_area_percentage(point.coords[0], forests, buffer_radius)
)
```

```python
# Assuming you have a 'forest_percentage' column in your 'points' GeoDataFrame
forest_percentages = points['forest_percentage'].to_numpy()

# Flatten and remove NaN values
forest_percentages = forest_percentages[~np.isnan(forest_percentages)]

# Calculate quantiles
forest_quantiles = np.percentile(forest_percentages, [10, 20, 30, 40, 50, 60, 70, 80, 90])

# Print results
print(forest_quantiles)
```
